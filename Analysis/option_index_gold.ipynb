```python
import pickle
import os
import json
import re

directory = "/home/atuin/g102ea/shared/group_10/"
```

```python
file_name = "results/analysis/charades/charades_tb_0_1%_0.5/input_ids.json"
file_path = os.path.join(directory, file_name)

input_ids_list = []
try:
    with open(file_path, "rb") as f:
        input_ids_list = pickle.load(f)
except Exception as e:
    print(f"Error reading {file_name}: {e}")
# print(len(input_ids_list))
# print(input_ids_list)
```

```python
voc_file_name = "model_checkpoints/qwen2vl-7b-instruct/vocab.json"

voc_file_path = os.path.join(directory, voc_file_name)
try:
    with open(voc_file_path, "r", encoding="utf-8") as f:
        data = json.load(f)
except Exception as e:
    print(f"Error reading {voc_file_name}: {e}")

# print(json.dumps(data, indent=4, ensure_ascii=False))
```

```python
subset_charades_mcq = "datasets/charades/subset_charades_mcq.json"
scm_file_path = os.path.join(directory, subset_charades_mcq)
try:
    with open(scm_file_path, "r", encoding="utf-8") as f:
        data_scm = json.load(f)
except Exception as e:
    print(f"Error reading {voc_file_name}: {e}")


# print(data_scm['mcq_data'].keys())
# print(data_scm['mcq_data']['mcqs'][0])
# print(data_scm['mcq_data']['mcq_labels'][0])
```

```python
id_to_token = {v: k for k, v in data.items()}
# print(json.dumps(id_to_token, indent=4, ensure_ascii=False))
```

```python
tokenized_input = []
for sublist in input_ids_list:
    tokenized_input.append([id_to_token.get(id) for id in sublist])

# print(tokenized_input[0])
```

```python
mcq_questions = data_scm['mcq_data']['mcqs']
mcq_answers = data_scm['mcq_data']['mcq_labels']  

options_list = []

for question, answer in zip(mcq_questions, mcq_answers):
    option = next(f"{i}" for i, t in re.findall(r"\((\d+)\) (.+)", question) if t.strip() == answer.strip())

    
    options_list.append(option)

# print(options_list)
```

```python
# print(tokenized_input[0])
# print(options_list[0])
```

```python
option_indices_list = []

for i, (tokens, option) in enumerate(zip(tokenized_input, options_list)):
    for idx in range(len(tokens) - 2):
        if tokens[idx] == "(" and tokens[idx + 1] == option and tokens[idx + 2] == ")":
            start_idx = idx
            break
    end_idx = start_idx + 3 
    while end_idx < len(tokens) and tokens[end_idx] != 'Ċ':
        end_idx += 1

    option_indices_list.append((start_idx, end_idx))


# print(option_indices_list)
# print(len(option_indices_list))
```
